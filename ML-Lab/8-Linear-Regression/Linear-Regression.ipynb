```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
df = pd.read_csv('../datasets/iris.csv')

newdf = df[['petal.length','petal.width']]
train_x, test_x , train_y, test_y= train_test_split(newdf['petal.length'],newdf['petal.width'],test_size=0.3,random_state=3, shuffle=True)
train_x = np.array(train_x).reshape(-1,1)
test_x = np.array(test_x).reshape(-1,1)
```

```python
lr = linear_model.LinearRegression()
lr.fit(train_x,train_y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(test_x,test_y)
y_predict = lr.predict(test_x)
plt.scatter(test_x,y_predict,color = 'red')
plt.xlabel("petal length")
plt.ylabel("petal width")
```

```python
import sklearn.metrics as skm

print('Root mean Square Error ', skm.mean_squared_error(test_y,y_predict)**0.5)
print('Mean absolute error ',skm.mean_absolute_error(y_predict,test_y))
```

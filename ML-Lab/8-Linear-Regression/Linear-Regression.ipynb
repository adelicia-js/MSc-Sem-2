```python
# Import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Importing a dataset 'iris.csv'
data = pd.read_csv('../datasets/iris.csv')
# data.head()
data.columns
```

```python
# Plotting a heatmap to visualize correlation between different attributes
sns.heatmap(data.corr(numeric_only=True),annot=True)
```

```python
# Splitting dataset into dependent and independent variables

# Independent variable 
X=data['petal.length']

# Dependent variable
y=data['petal.width']
```

```python
# Splitting dataset into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=1,shuffle=True)

# Reshaping the nd arrays
X_train = np.array(X_train).reshape(-1,1)
# X_train
X_test = np.array(X_test).reshape(-1,1)
# X_test
```

```python
# Creating a model
model = LinearRegression()

# Training the model
model.fit(X_train,y_train)

# Forming predictions
y_pred = model.predict(X_test)
```

```python
# Plotting the regression line
plt.scatter(X_train,y_train)
plt.plot(X_test,y_pred,color="red")
```

```python
# Intercept of regression line
print("Intercept of model:",model.intercept_)

# Coefficient of regression line
print("Coefficient of the model",model.coef_)
```

```python
# Regression Line
print("Regression line is\ny=",model.intercept_,"+",model.coef_[0],"x")
```

```python
# Mean Squared Error
print("Mean squared error is",mean_squared_error(y_test,y_pred))
```

```python
# Root Mean Square Error
print("The root mean square error is",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
# Mean Absolute Error
print("Mean absolute error is",mean_absolute_error(y_test,y_pred))
```

```python
# R Squared Error
print("R square error is",r2_score(y_test,y_pred))
```

```python
# Accuracy of model in %
print("The accuracy of the model is",(round(r2_score(y_test,y_pred),4)*100),"%")
```
